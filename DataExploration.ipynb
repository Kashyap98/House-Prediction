```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
house_prices = pd.read_csv('train.csv')
print(house_prices.columns)
```

```python
def count_unique(house_prices, cols):
    for col in cols:
        print('\n' + 'For column ' + col)
        print(house_prices[col].value_counts())

#Get the unique values for each column. This is helpful to see if there is a bias, or if there is not enough variation to be
#helpful in a potential prediction
cat_cols = ['LotFrontage']
count_unique(house_prices, cat_cols)
```

```python
def plot_bars(house_prices, cols):
    for col in cols:
        fig = plt.figure(figsize=(6,6)) # define plot area
        ax = fig.gca() # define axis    
        counts = house_prices[col].value_counts() # find the counts for each unique category
        counts.plot.bar(ax = ax, color = 'blue') # Use the plot.bar method on the counts data frame
        ax.set_title('Number of houses by ' + col) # Give the plot a main title
        ax.set_xlabel(col) # Set text for the x axis
        ax.set_ylabel('Number of houses')# Set text for y axis
        plt.show()

##This is a barchart visualization the previous section. 
plot_cols = ['OverallQual']
plot_bars(house_prices, plot_cols)    
```

```python
def plot_histogram(house_prices, cols, bins = 10):
    for col in cols:
        fig = plt.figure(figsize=(6,6)) # define plot area
        ax = fig.gca() # define axis    
        house_prices[col].plot.hist(ax = ax, bins = bins) # Use the plot.hist method on subset of the data frame
        ax.set_title('Histogram of ' + col) # Give the plot a main title
        ax.set_xlabel(col) # Set text for the x axis
        ax.set_ylabel('Number of houses')# Set text for y axis
        plt.show()
        
num_cols = ['OverallQual']    
plot_histogram(house_prices, num_cols)
```

```python
def plot_scatter(house_prices, cols, col_y = 'SalePrice'):
    for col in cols:
        fig = plt.figure(figsize=(7,6)) # define plot area
        ax = fig.gca() # define axis   
        house_prices.plot.scatter(x = col, y = col_y, ax = ax)
        ax.set_title('Scatter plot of ' + col_y + ' vs. ' + col) # Give the plot a main title
        ax.set_xlabel(col) # Set text for the x axis
        ax.set_ylabel(col_y
                     )# Set text for y axis
        plt.show()

##Works better with numerical columns. But creates a scatterplot.
num_cols = ['OverallQual']
plot_scatter(house_prices, num_cols)  
```

```python
def plot_box(house_prices, cols, col_y = 'SalePrice'):
    for col in cols:
        sns.set_style("whitegrid")
        sns.boxplot(col, col_y, data=house_prices)
        plt.xlabel(col) # Set text for the x axis
        plt.ylabel(col_y)# Set text for y axis
        plt.show()
        
cat_cols = ['Neighborhood']
plot_box(house_prices, cat_cols)    
```

```python
##shape_col is for determining which category the shapes are based off of.
def plot_scatter_shape(house_prices, cols, shape_col = 'LotShape', col_y = 'SalePrice', alpha = 0.2):
    shapes = ['+', 'o', 's', 'x', '^'] # pick distinctive shapes
    unique_cats = house_prices[shape_col].unique()
    for col in cols: # loop over the columns to plot
        sns.set_style("whitegrid")
        for i, cat in enumerate(unique_cats): # loop over the unique categories
            temp = house_prices[house_prices[shape_col] == cat]
            sns.regplot(col, col_y, data=temp, marker = shapes[i], label = cat,
                        scatter_kws={"alpha":alpha}, fit_reg = False, color = 'blue')
        plt.title('Scatter plot of ' + col_y + ' vs. ' + col) # Give the plot a main title
        plt.xlabel(col) # Set text for the x axis
        plt.ylabel(col_y)# Set text for y axis
        plt.legend()
        plt.show()
            
num_cols = ['1stFlrSF']
plot_scatter_shape(house_prices, num_cols)  
```
